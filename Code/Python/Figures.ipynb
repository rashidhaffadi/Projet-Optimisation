```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

%matplotlib inline
np.random.seed(0)
```

```python
N = 200
```

```python
x = np.linspace(-3.2, 3.2, N)
y = np.cos(2*x)
b = np.random.normal(0, 0.2, y.shape)
f = y + b
```

```python
plt.plot(x, y, color = 'r', label = 'cos(2x)')
plt.scatter(x, f, color = 'b', label = 'cos(2x) + bruit')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper right')
plt.title('Figure 1 : fonction cos(2x) + bruit')
plt.show()
```

```python
def moment(mu):
    m = []
    m.append(0)
    for i in range (1,N):
        m.append((mu*m[i-1] + (1-mu)*f[i]))
    return m
m = moment(0.9)
```

```python
plt.plot(x, y, color = 'r', label = 'cos(2x)')
plt.plot(x, m, color = 'y', label = 'moyennes pondérées mu=0.9')
plt.scatter(x, f, color = 'b', label = 'cos(2x) + bruit')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper right')
plt.title('Figure 2 : fonction cos(2x) + bruit avec momentum')
plt.show()
```

```python
m1 = moment(0.5)
m2 = moment(0.98)
```

```python
plt.plot(x, y, color = 'r', label = 'cos(2x)')
plt.plot(x, m1, color = 'c', label = 'moyennes pondérées mu=0.5')
plt.plot(x, m, color = 'g', label = 'moyennes pondérées mu=0.9')
plt.plot(x, m2, color = 'y', label = 'moyennes pondérées mu=0.98')
plt.scatter(x, f, color = 'b', label = 'cos(2x) + bruit')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper right')
plt.title('Figure 3 :moyennes pondérées avec mu différentes ')
plt.show()
```
